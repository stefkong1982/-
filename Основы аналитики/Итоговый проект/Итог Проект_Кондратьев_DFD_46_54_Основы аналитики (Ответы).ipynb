```python
# создаю датафрейм с названием города и его населением из таблицы City
df_population = df [['name', 'population']]. copy ()
df_population.head(15)
```

```python
# соединяю данные из выгруженного своей таблички и таблицой с населением города.
df_2 = df_population.merge(df_otpusk, 
 left_on ='name', 
right_on ='name', 
     how ='outer')
df_2.head(10)
```

```python
# создаю датафрейм с отсортированый только по моим городам
df_3 = df_2[df_2['name'].isin(['Ibiza', 'Goa', 'Bali', 'Phuket', 'Sochi', 'Gelendzhik', 'Anapa', 'Rio de Janeiro', 'Miami', 'Dubai',])]
df_3.head(15)
```
