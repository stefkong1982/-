## Задание 1
Дана переменная, в которой хранится словарь, содержащий гео-метки для каждого пользователя (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран множество уникальных гео-меток всех пользователей.

Пример работы программы:  

```
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}
```

Результат:
`{98, 35, 15, 213, 54, 119}`

### Решение Задания 1

Для решения данной задачи можно использовать встроенный метод `set()`, который позволяет создавать множество уникальных элементов.

````python
# @title Код решения
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}

unique_geo_marks = set()

for geo_marks in ids.values():
    unique_geo_marks.update(geo_marks)

print(unique_geo_marks)
````

**Описания решения**

Создаем пустое множество `unique_geo_marks`. Затем, с помощью цикла `for`, мы перебираем значения в словаре `ids.values()`, которые представляют собой списки гео-меток для каждого пользователя. С помощью метода `update()` мы добавляем все значения в множество `unique_geo_marks`. В конце мы выводим полученное множество, которое будет содержать только уникальные гео-метки всех пользователей.


## Задание 2
Дана переменная, в которой хранится список поисковых запросов пользователя (пример структуры данных приведен ниже, но запросы потенциально могут быть любые). Вам необходимо написать программу, которая выведет на экран распределение количества слов в запросах в требуемом виде.

Пример работы программы:  

```
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
```

Результат:
```
Поисковых запросов, содержащих 2 слов(а): 42.86%
Поисковых запросов, содержащих 3 слов(а): 57.14%
```

### Решение Задания 2

Для решения данной задачи нам необходимо подсчитать количество слов в каждом запросе и определить распределение количества слов.

````python
# @title Код решения
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

word_count = {} # словарь для подсчета количества слов

# Перебираем каждый запрос в списке queries
for query in queries:
    words = query.split()  # разделяем запрос на отдельные слова
    count = len(words)  # подсчитываем количество слов
    if count in word_count:  # если данное количество слов уже есть в словаре
        word_count[count] += 1  # увеличиваем счетчик на 1
    else:
        word_count[count] = 1  # иначе создаем новую запись в словаре со счетчиком равным 1

total_queries = len(queries)  # общее количество запросов

# Выводим распределение количества слов
for count, frequency in word_count.items():
    percentage = (frequency / total_queries) * 100
    print(f"Поисковых запросов, содержащих {count} слов(а): {percentage:.2f}%")
````

**Описание решения**

В этом примере мы используем словарь `word_count`, чтобы подсчитать количество слов в каждом запросе. Мы перебираем каждый запрос в списке `queries`, разделяем его на отдельные слова с помощью метода `split()` и подсчитываем количество слов с помощью функции `len()`. Затем мы обновляем словарь `word_count` с использованием этого количества слов.

Затем мы вычисляем общее количество запросов в переменной `total_queries`.

Для вывода распределения количества слов в запросах мы используем цикл `for`, чтобы перебрать каждую пару ключ-значение в словаре `word_count`.

## Задание 3
Дана переменная, в которой хранится информация о затратах и доходе рекламных кампаний по различным источникам. Необходимо дополнить исходную структуру показателем [ROI](https://ru.wikipedia.org/wiki/%D0%9E%D0%BA%D1%83%D0%BF%D0%B0%D0%B5%D0%BC%D0%BE%D1%81%D1%82%D1%8C_%D0%B8%D0%BD%D0%B2%D0%B5%D1%81%D1%82%D0%B8%D1%86%D0%B8%D0%B9), который рассчитаем по формуле: **((revenue / cost) - 1) * 100**

Пример работы программы:  

```
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}
```

Результат:
```
{'adwords': {'revenue': 35, 'cost': 34, 'ROI': 2.94},
 'facebook': {'revenue': 103, 'cost': 110, 'ROI': -6.36},
 'twitter': {'revenue': 11, 'cost': 24, 'ROI': -54.17},
 'vk': {'revenue': 103, 'cost': 98, 'ROI': 5.1},
 'yandex': {'revenue': 179, 'cost': 153, 'ROI': 16.99}}
```

### Решение Задания 3
Для решения данной задачи, мы можем использовать цикл `for` для итерации по каждому источнику рекламы в словаре `results`. Затем, мы рассчитаем ROI для каждого источника согласно формуле ((revenue / cost) - 1) * 100 и добавим его в словарь для каждого источника.


````python
# @title Код решения
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}

for source, data in results.items():
    revenue = data['revenue']
    cost = data['cost']
    roi = ((revenue / cost) - 1) * 100
    data['ROI'] = round(roi, 2)

for source, data in results.items():
    print(f"{source}: {data}")
````

**Описание решения**

- `source` - переменная, в которую будет сохранен ключ элемента словаря (в данном случае, это источник рекламы, например, 'vk', 'yandex', 'facebook' и т.д.).
- `data` - переменная, в которую будет сохранено значение элемента словаря (в данном случае, это словарь с информацией о доходе и затратах для данного источника рекламы).

Внутри цикла мы получаем значения дохода (`revenue`) и затрат (`cost`) из словаря `data` с помощью обращения по ключам `'revenue'` и `'cost'`.

Затем мы рассчитываем показатель ROI (Return on Investment) с помощью формулы ((revenue / cost) - 1) * 100 и сохраняем его в переменной `roi`.

Для сохранения значения ROI в словаре `data`, мы добавляем новый ключ `'ROI'` с помощью выражения `data['ROI']`. Значение `round(roi, 2)` округляется до двух десятичных знаков с помощью функции `round()` и сохраняется под ключом `'ROI'`

## Задание 4
Дана переменная, в которой хранится статистика рекламных каналов по объемам продаж (пример структуры данных приведен ниже). Напишите программу, которая возвращает название канала с максимальным объемом продаж.

Пример работы программы:  

`stats = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}`

Результат:
`Максимальный объем продаж на рекламном канале: vk`

### Решение Задания 4

Для решения данной задачи, мы можем использовать функцию `max()`, чтобы найти максимальное значение в словаре и вернуть соответствующий ключ.

````python
# @title Код решения
stats = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}

max_channel = max(stats, key=stats.get)

print("Максимальный объем продаж на рекламном канале:", max_channel)
````

**Описание решения**

`max()` - это встроенная функция Python, которая принимает итерируемый объект и возвращает его максимальный элемент. Однако, вместо того, чтобы сравнивать элементы напрямую, мы можем указать функцию `key`, которая будет вызываться для каждого элемента и возвращать значение, по которому будет происходить сравнение.

В нашем случае `key=stats.get` указывает функции `max()` использовать метод `.get()` словаря `stats` для извлечения значения по ключу. Функция `stats.get` возвращает значение для каждого ключа в словаре, и на основе этих значений будет происходить сравнение элементов в функции `max()`. Таким образом, `max(stats, key=stats.get)` вернет ключ с максимальным значением в словаре `stats`.


## Задание 5 (необязательно)
Дан список произвольной длины. Необходимо написать код, который на основе исходного списка составит словарь такого уровня вложенности, какова длина исхондого списка.

Примеры работы программы:  

1. `my_list = ['2018-01-01', 'yandex', 'cpc', 100]`

Результат:
`{'2018-01-01': {'yandex': {'cpc': 100}}}`

2. `my_list = ['a', 'b', 'c', 'd', 'e', 'f']`

Результат:
`{'a': {'b': {'c': {'d': {'e': 'f'}}}}}`

### Решение Задания 5

Для решения этой задачи. Мы можем использовать цикл и итерироваться по списку в обратном порядке, чтобы создавать вложенные словари.


````python
# @title Код решения
my_list = ['2018-01-01', 'yandex', 'cpc', 100]
result = my_list[-1]  # Инициализируем результат последним элементом списка

for item in reversed(my_list[:-1]):
    result = {item: result}

print(result)
````

**Описание решения**

Функция `reversed()` возвращает итератор, который обходит последовательность в обратном порядке.

Строка `result = {item: result}` создает новый словарь, где `item` является ключом, а значение равно текущему значению `result`.

При каждой итерации цикла в данной строке мы создаем новый словарь с одной парой ключ-значение. Ключом является значение `item` - элемент текущей итерации цикла, а значением является текущее значение переменной `result`.

На каждой итерации создается новый словарь, где предыдущий словарь `result` становится значением нового ключа `item`. Это позволяет нам постепенно собирать вложенный словарь, где каждый новый ключ становится вложенным словарем внутри предыдущего словаря.

### Решение Задания 6

````python
# @title Код решения
cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},
    ]
}

portions = int(input('Введите количество порций:\n'))

summary = {}

for dish in cook_book.values():
    for ingredient in dish:
        key = (ingredient['ingridient_name'], ingredient['measure'])
        if key not in summary:
            summary[key] = ingredient['quantity'] * portions
        else:
            summary[key] += ingredient['quantity'] * portions

for key, val in summary.items():
    print(f'{key[0].capitalize()}: {val} {key[1]}')
````

**Описание решения**

2. Пользователю предлагается ввести количество порций с помощью функции `input()`. Введенное значение сохраняется в переменной `portions`.

3. Создается пустой словарь `summary`, который будет использоваться для подсчета и хранения суммарных значений ингредиентов.

4. Происходит итерация по всем значениям (спискам ингредиентов) словаря `cook_book`, используя цикл `for dish in cook_book.values()`. Внутри этого цикла происходит еще одна итерация по ингредиентам каждого блюда, используя цикл `for ingredient in dish`.

5. Для текущего ингредиента формируется кортеж `key`, содержащий название ингредиента и его единицу измерения. Этот кортеж будет использоваться в качестве ключа для словаря `summary`.

6. Если ингредиента с таким ключом еще нет в словаре `summary`, то его значение инициализируется суммарным количеством данного ингредиента, умноженным на количество порций (`ingredient['quantity'] * portions`). Если ингредиент с таким ключом уже присутствует в словаре `summary`, то его значение увеличивается на суммарное количество данного ингредиента, умноженное на количество порций.

7. После окончания циклов происходит итерация по парам ключ-значение в словаре `summary` с помощью цикла `for key, val in summary.items()`.

8. Для каждой пары ключ-значение выводится информация о количестве ингредиента с его названием и единицей измерения, используя форматированную строку `print(f'{key[0].capitalize()}: {val} {key[1]}')`. Метод `capitalize()` используется для названия ингредиента, чтобы первая буква была заглавной.
