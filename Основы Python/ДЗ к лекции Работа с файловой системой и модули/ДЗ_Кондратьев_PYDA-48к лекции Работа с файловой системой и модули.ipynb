## Задание 1

Переведите содержимое файла purchase_log.txt* в словарь purchases вида:
```
{‘1840e0b9d4’: ‘Продукты’, …}
```



### Решение задания 1

````python
#Смотрим что в файле purchase_log.txt
with open('purchase_log.txt', 'r') as file:
   print([next(file) for x in range(5)])
````

````python
# В этом коде мы считываем файл `purchase_log.txt` построчно и создаем словарь `purchases`, в котором ключами являются `user_id`, а значениями - `category`.
import json

purchases = {}

with open('purchase_log.txt', 'r') as file:
    next(file)

    for line in file:
        line = json.loads(line.strip())

        user_id = line['user_id']
        purchases[user_id] = line['category']

for key in list(purchases.keys())[:5]:
    print(f'{key}: {purchases[key]}')
````

**Описание работы кода:**



1. Сначала мы импортируем модуль `json`, чтобы работать с данными в формате JSON.
2. Создаем пустой словарь `purchases`, в котором будем хранить информацию о покупках.
3. Открываем файл `purchase_log.txt` в режиме чтения (`'r'`) с помощью оператора `open()`.
4. Считываем первую строку файла, используя функцию `next(file)`. Обычно первая строка содержит заголовки столбцов, и в данном случае мы просто пропускаем эту строку.
5. Затем переходим к обработке остальных строк файла. Для каждой строки:
    - С помощью `json.loads(line.strip())` мы преобразуем строку в словарь, используя функцию `json.loads()` Символ новой строки (`\n`) `strip()`.
    - Извлекаем значения `user_id` и `category` из словаря `line`.
    - Добавляем `user_id` в качестве ключа и `category` в качестве значения в словарь `purchases`.
6. После обработки всех строк мы выводим первые 5 элементов словаря `purchases` с помощью цикла `for key in list(purchases.keys())[:5]`, где `list(purchases.keys())` возвращает список ключей словаря, а `[:5]` берет первые 5 ключей из списка. Каждый элемент выводится в формате `'{key}: {purchases[key]}'`, где `key` - ключ словаря, а `purchases[key]` - соответствующее значение..


## Задание 2

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки, если покупка была, сам файл visit_log.csv* изменять не надо. Запишите в файл funnel.csv визиты из файла visit_log.csv*, в которых были покупки с указанием категории.

### Решение задания 2

````python
# Смотрим что в файле visit_log.csv
with open('visit_log.csv', 'r') as f:
  print([next(f) for x in range(5)])
````

````python
# В этом коде мы открываем файл `visit_log.csv` для чтения и создаем файл `funnel.csv` для записи.
# Затем мы последовательно обрабатываем каждую строку из файла `visit_log.csv` и проверяем, если `user_id`
# содержится в словаре `purchases` (который мы создавали в первом задании),
# то записываем соответствующую строку в файл `funnel.csv` с добавлением категории покупки.
with open('visit_log.csv', 'r') as f:
    with open('funnel.csv', 'w') as funnel:
        for line in f:
            user_id, source = line.strip().split(',')

            if user_id in purchases:
                funnel.write(f'{user_id},{source},{purchases[user_id]}\n')

with open('funnel.csv', 'r') as f:
    print([next(f) for _ in range(5)])
````

**Описание работы кода:**



1. С помощью оператора `with` открываем файл `visit_log.csv` в режиме чтения (`'r'`). Этот оператор позволяет автоматически закрыть файл после его использования.
2. Вложенный оператор `with` открывает файл `funnel.csv` в режиме записи (`'w'`).
3. Затем мы переходим к обработке каждой строки в файле `visit_log.csv`. Для каждой строки:
    - С помощью `line.strip().split(',')` мы удаляем Символ новой строки (`\n`) (`strip()`) и разделяем строку на поля, используя запятую как разделитель (`split(',')`). Получаем две переменные `user_id` и `source`.
    - Затем мы проверяем, если `user_id` содержится в словаре `purchases`, используя оператор `if user_id in purchases:`. Если условие выполняется, то продолжаем следующие шаги.
    - Мы записываем строку в файл `funnel.csv` с помощью метода `write()`, формируя строку вида `'{user_id},{source},{purchases[user_id]}\n'`, где `user_id` и `source` - это значения из файл `visit_log.csv`, а `purchases[user_id]` - соответствующая категория покупки из словаря `purchases`.
4. После обработки всех строк в файле `visit_log.csv` закрываем файлы с помощью операторов `with`.

Затем мы снова открываем файл `funnel.csv`, который мы только что создали, и выводим первые 5 строк файла с помощью `[next(f) for _ in range(5)]`, где `next(f)` читает следующую строку из файла, а `_` используется как пустая переменная, так как в данном случае нас интересует только количество строк, а не их содержимое.
