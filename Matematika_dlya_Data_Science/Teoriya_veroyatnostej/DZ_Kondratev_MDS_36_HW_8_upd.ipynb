### Задание 1

В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

Вероятность того, что случайно купленная новая продукция окажется высшего сорта, можно найти путем умножения вероятности первого события (выбрать продукцию одного из предприятий) на вероятность второго события (выбрать продукцию высшего сорта на выбранном предприятии).

* Вероятность выбрать продукцию первого предприятия - 20% или 0.2.
* Вероятность выбрать продукцию второго предприятия - 30% или 0.3.
* Вероятность выбрать продукцию третьего предприятия - 50% или 0.5.



* Вероятность выбрать продукцию высшего сорта на первом предприятии - 10% или 0.1.
* Вероятность выбрать продукцию высшего сорта на втором предприятии - 5% или 0.05.
* Вероятность выбрать продукцию высшего сорта на третьем предприятии - 20% или 0.2.

Вероятность того, что случайно купленная новая продукция окажется высшего сорта:

(0.2 * 0.1) + (0.3 * 0.05) + (0.5 * 0.2) = 0.02 + 0.015 + 0.1 = 0.135

Ответ: Вероятность того, что случайно купленная новая продукция окажется высшего сорта, равна 0.135 или 13.5%.

```
# Вычисляем аналитически

# Задаем вероятности выбора продукции с разных предприятий
p1 = 0.2 # вероятность выбора продукции первого предприятия
p2 = 0.3 # вероятность выбора продукции второго предприятия
p3 = 0.5 # вероятность выбора продукции третьего предприятия

# Задаем вероятности выбора продукции высшего сорта на разных предприятиях
p1_high = 0.1 # вероятность выбора продукции высшего сорта на первом предприятии
p2_high = 0.05 # вероятность выбора продукции высшего сорта на втором предприятии
p3_high = 0.2 # вероятность выбора продукции высшего сорта на третьем предприятии

# Находим вероятность того, что случайно купленная продукция окажется высшего сорта
probability = (p1 * p1_high) + (p2 * p2_high) + (p3 * p3_high)

# Выводим результат
print("Вероятность того, что случайно купленная новая продукция окажется высшего сорта: ", probability)
```

```
# Получаем с помощью моделирования случайных величин

import numpy as np
from scipy import stats

# Задаем вероятности выбора продукции с разных предприятий
p1 = 0.2 # вероятность выбора продукции первого предприятия
p2 = 0.3 # вероятность выбора продукции второго предприятия
p3 = 0.5 # вероятность выбора продукции третьего предприятия

# Задаем вероятности выбора продукции высшего сорта на разных предприятиях
p1_high = 0.1 # вероятность выбора продукции высшего сорта на первом предприятии
p2_high = 0.05 # вероятность выбора продукции высшего сорта на втором предприятии
p3_high = 0.2 # вероятность выбора продукции высшего сорта на третьем предприятии

# Создаем дискретные случайные величины для выбора продукции
rv1 = stats.bernoulli(p1)
rv2 = stats.bernoulli(p2)
rv3 = stats.bernoulli(p3)

# Создаем дискретные случайные величины для выбора продукции высшего сорта
rv1_high = stats.bernoulli(p1_high)
rv2_high = stats.bernoulli(p2_high)
rv3_high = stats.bernoulli(p3_high)

# Моделируем выбор продукции с помощью дискретных случайных величин
product1 = rv1.rvs(100000)
product2 = rv2.rvs(100000)
product3 = rv3.rvs(100000)

# Моделируем выбор продукции высшего сорта на каждом предприятии
product1_high = rv1_high.rvs(100000)
product2_high = rv2_high.rvs(100000)
product3_high = rv3_high.rvs(100000)

# Вычисляем вероятность того, что случайно купленная продукция окажется высшего сорта
probability = np.mean((product1 * product1_high) + (product2 * product2_high) + (product3 * product3_high))

# Выводим результат
print("Вероятность того, что случайно купленная новая продукция окажется высшего сорта: ", probability)
```

В обоих версиях кода была вычислена вероятность того, что случайно купленная новая продукция окажется высшего сорта. В первом коде вероятность была вычислена аналитически, а во втором коде вероятность была получена с помощью моделирования случайных величин.

Полученные результаты практически идентичны: вероятность составляет около 0.135 в первом коде и 0.13486 во втором коде.

Вывод: вероятность того, что случайно купленная новая продукция окажется высшего сорта, примерно равна 0.135.

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

События А1, А2 и А3 попарно независимы, потому что наступление любого из этих событий не влияет на вероятность наступления двух других. Детальнее рассмотрим это утверждение.

1) Для событий А1 и А2: наступление события А1 (монеты 1 и 2 упали одной стороной) не меняет вероятности наступления события А2 (монеты 2 и 3 упали одной стороной), которая составляет 1/2 (так как монета находится в равновесии).

2) Для событий А1 и А3: аналогично, наступление события А1 не меняет вероятности наступления события А3.

3) Для событий А2 и А3: наступление события А2 не меняет вероятности наступления события А3.

Таким образом, все три события попарно независимы.

Однако, если рассмотреть все три события вместе, они становятся зависимыми.

В том случае, если два из трех событий произошли (скажем, А1 и А2), то третье событие (А3) произойдет автоматически.

Чтобы понять почему, рассмотрим ситуацию: если монеты 1 и 2 упали одной стороной (событие А1) и монеты 2 и 3 упали одной стороной (событие А2),

это автоматически означает, что монеты 1 и 3 упали одной стороной (событие А3),

так как обе они упали так же как и монета номер 2.

Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, t
```

```
# Параметры для нормального распределения
mean = [2, 2, 2, 2, 2]
std = [0.25, 0.50, 0.75, 1, 2]

# Построение графиков плотности для нормального распределения
plt.figure(figsize=(10, 6))
for i in range(len(mean)):
    x = np.linspace(-5, 10, 100)
    y = norm.pdf(x, mean[i], std[i])
    plt.plot(x, y, label=f'mean={mean[i]}, std={std[i]}')
plt.title('Нормальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

1. Нормальное распределение:
   - Параметры: среднее значение (`mean`) и стандартное отклонение (`std`).
   - Большее значение среднего значения (`mean`) смещает график плотности вправо, меньшее значение смещает его влево.
   - Увеличение значения стандартного отклонения (`std`) делает график плотности шире и более полого, уменьшение значения делает его уже и `острее`.

```
# Параметры для экспоненциального распределения
lambda_ = [0.25, 0.5, 0.75, 1, 2]

# Построение графиков плотности для экспоненциального распределения
plt.figure(figsize=(10, 6))
for i in range(len(lambda_)):
    x = np.linspace(0, 6, 100)
    y = expon.pdf(x, scale=1/lambda_[i])
    plt.plot(x, y, label=f'lambda={lambda_[i]}')
plt.title('Экспоненциальное распределение')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

2. Экспоненциальное распределение:
   - Параметр: `lambda` (1/среднее значение)
   - Увеличение значения `lambda` делает график плотности более пологим на левой стороне и `более тяжелым` на правой стороне.
   - Уменьшение значения `lambda` делает график плотности более пологим на правой стороне и `более тяжелым` на левой стороне.

```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# Параметры для распределения Стьюдента
df = [1, 2, 5, 10, 30]

# Построение графиков плотности для распределения Стьюдента
plt.figure(figsize=(10, 6))
for i in range(len(df)):
    x = np.linspace(-5, 5, 100)
    y = t.pdf(x, df[i])
    plt.plot(x, y, label=f'df={df[i]}')
plt.title('Распределение Стьюдента')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

3. Распределение Стьюдента:
   - Параметр: степени свободы (`df`)
   - Увеличение значения степеней свободы делает график плотности более заостренным вокруг среднего значения.
   - Уменьшение значения степеней свободы делает график плотности более выпуклым и `тяжелым` в хвостах.
