### Решение Задания 2

В григорианском календаре обычный год состоит из 365 дней. Поскольку фактическая продолжительность звездного года (период обращения Земли вокруг Солнца) составляет 365.2425 дней, «високосный год» в 366 дней используется один раз каждые четыре года для устранения ошибки, вызванной тремя обычными (но короткими) годами. Любой год, который делится на 4 без остатка, является високосным годом: например, 1988, 1992 и 1996 годы являются високосными годами.

Тем не менее, есть еще небольшая ошибка, которая должна быть учтена. Чтобы устранить эту ошибку, григорианский календарь предусматривает, что год, который делится без остатка на 100 (например, 1900) является високосным годом только в том случае, если он также без остатка делится на 400.

Високосный год выпадает каждые 4 года, но каждый 100-й год не является високосным, кроме каждого 400-го года.

По этой причине следующие годы не являются високосными:

1700, 1800, 1900, 2100, 2200, 2300, 2500, 2600


Это потому, что они делятся без остатка на 100, но не на 400.

Следующие годы – високосные: 1600, 2000, 2400

Это потому, что они делятся без остатка на 100 и 400.


**Как определить, является ли год високосным**

Чтобы определить, является ли год високосным, выполним следующие действия:

1. Если год делится на 4 без остатка, перейдём на Шаг-2. В противном случае
перейдём на Шаг-5.
2. Если год делится на 100 без остатка, перейдём на Шаг-3. В противном случае перейдём на Шаг-4.
3. Если год делится на 400 без остатка, перейдём на Шаг-4. В противном случае перейдём на Шаг-5.
4. Год високосный (366 дней).
5. Год не високосный год (365 дней).


Для выполнения данных шагов будем использовать конструкциию `if-elif-else`

Примеры радоты программы рассмотрим ниже.


````python
# @title Пример №1 - проверяем является ли 1900 високосным
year_1 = 1900 # Здесь мы просто определяем переменную `year_1` и присваиваем ей значение 1900. Следующие операции будут выполняться на основании этого значения.
if year_1 % 4 == 0: # Шаг-1 - Если год делится на 4 без остатка, перейдём к Шагу 2. В противном случае перейдём к Шагу 5 - print("Обычный год").
    if year_1 % 100 == 0: # Шаг-2 - Если год делится на 100 без остатка, перейдём к Шагу 3. В противном случае перейдём к Шагу 4 - print("Високосный год").
        if year_1 % 400 == 0: # Шаг-3 - Если год делится на 400 без остатка, перейдём к Шагу 4 - print("Високосный год"),
            print("Високосный год")     # В противном случае перейдём на Шаг-5. - print("Обычный год").
        else:
            print("Обычный год")
    else:
        print("Високосный год")
else:
    print("Обычный год")
````

````python
# @title Пример №2 - проверяем является ли 2000 високосным
year_1 = 2000 # Здесь мы просто определяем переменную `year_1` и присваиваем ей значение 2000. Следующие операции будут выполняться на основании этого значения.
if year_1 % 4 == 0: # Шаг-1 - Если год делится на 4 без остатка, перейдём к Шагу 2. В противном случае перейдём к Шагу 5 - print("Обычный год").
    if year_1 % 100 == 0: # Шаг-2 - Если год делится на 100 без остатка, перейдём к Шагу 3. В противном случае перейдём к Шагу 4 - print("Високосный год").
        if year_1 % 400 == 0: # Шаг-3 - Если год делится на 400 без остатка, перейдём к Шагу 4 - print("Високосный год"),
            print("Високосный год")     # В противном случае перейдём на Шаг-5. - print("Обычный год").
        else:
            print("Обычный год")
    else:
        print("Високосный год")
else:
    print("Обычный год")
````

**Описание работы кода:**

1. В начале объявляется переменная `year_1`, которая содержит значение 2000 (наш год).
2. Далее идет проверка условия `if year_1 % 4 == 0` - это первый шаг. Если год делится на 4 без остатка, то выполняются дальнейшие шаги. Если год не делится на 4 без остатка, то сразу же выводится "Обычный год".
3. Если первое условие выполнено, то внутри первого условия идет проверка условия `if year_1 % 100 == 0` - это второй шаг. Если год делится на 100 без остатка, то выполняются следующие шаги. Если год не делится на 100 без остатка, то сразу же выводится "Високосный год".
4. Если второе условие выполнено, то внутри второго условия идет проверка условия `if year_1 % 400 == 0` - это третий шаг. Если год делится на 400 без остатка, то выводится "Високосный год". Если год не делится на 400 без остатка, то выводится "Обычный год".
5. Если второе условие не выполнено, то выводится "Високосный год".
6. Если первое условие не выполнено, то выводится "Обычный год".

Таким образом, код проверяет заданное условие для определения, является ли год високосным или обычным, и выводит соответствующий результат.

Посмотрим на примере с годом 2000, который вы использовали.

1. Сначала выполняется первое условие `if year_1 % 4 == 0`. Год 2000 делится на 4 без остатка, поэтому переходим к следующему шагу.
2. Затем выполняется второе условие `if year_1 % 100 == 0`. Год 2000 также делится на 100 без остатка, поэтому переходим к следующему шагу.
3. И наконец, выполняется третье условие `if year_1 % 400 == 0`. Год 2000 также делится на 400 без остатка, поэтому выводится "Високосный год".

Таким образом, исходя из условий в коде, год 2000 является високосным.

### Решение Задания 4

совместсно с операторами `and и or`

````python
# Запрашиваем у пользователя ширину, длину и высоту товара
width = int(input("Введите ширину: "))
length = int(input("Введите длину: "))
height = int(input("Введите высоту: "))

# Проверяем соответствие размеров товара требованиям
if width <= 15 and length <= 15 and height <= 15:
    # Если каждое измерение меньше или равно 15 сантиметрам, выводим "Коробка №1"
    print("Коробка №1")
elif width > 200 or length > 200 or height > 200:
    # Если хотя бы одно измерение больше 2 метров, выводим "Упаковка для лыж"
    print("Ищите упаковку для лыж")
elif width > 15 and width < 50 or length > 15 and length < 50 or height > 15 and height < 50:
    # Если хотя бы одно измерение больше 15 сантиметров, но меньше 50 сантиметров, выводим "Коробка №2"
    print("Коробка №2")
else:
    # Во всех остальных случаях выводим "Стандартная коробка №3"
    print("Стандартная коробка №3")
````

Данный код запрашивает у пользователя ввод размеров товара (ширины, длины и высоты в сантиметрах) и на основе этих данных определяет подходящую упаковку для товара.

## Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Несчастливый билет`  
  
2. ```number = 123321```

Результат:  
`Счастливый билет`


### Решение Задания 5

Для решения задачи будем использовать
1. Конструкцию `start:stop:step`
2. Цикл `for`

````python
number = input("Введите номер билета: ")  # Запрашиваем у пользователя ввод номера билета. Результат сохраняется в переменной "number"
first_three = number[:3]  # Выделяем первые три символа номера билета и сохраняем в переменной "first_three"
last_three = number[-3:]  # Выделяем последние три символа номера билета и сохраняем в переменной "last_three"
sum_of_first_three = sum(int(digit) for digit in first_three)  # Суммируем цифры первых трех символов номера билета и сохраняем \результат в переменной "sum_of_first_three"
sum_of_last_three = sum(int(digit) for digit in last_three)  # Суммируем цифры последних трех символов номера билета и сохраняем результат в переменной "sum_of_last_three"

if sum_of_first_three == sum_of_last_three:  # Проверяем, равны ли суммы первых и последних трех цифр номера билета
    print("Счастливый билет")  # Если суммы равны, выводим сообщение "Счастливый билет"
else:
    print("Несчастливый билет")  # Если суммы не равны, выводим сообщение "Несчастливый билет"
````

**Описание работы кода:**

1. Вводится шестизначное число (`number`), представляющее номер проездного билета.
2. Используется срез для получения первых трех цифр (`first_three`) и последних трех цифр (`last_three`) числа `number`.
3. Вычисляется сумма первых трех цифр (`sum_of_first_three`) с помощью генератора списка, где каждая цифра преобразуется в целое число.
4. Вычисляется сумма последних трех цифр (`sum_of_last_three`) аналогичным образом, только используется `last_three`.
5. Сравнивается `sum_of_first_three` с `sum_of_last_three`:
   - Если суммы совпадают, выводится сообщение "Счастливый билет".
   - Если суммы не совпадают, выводится сообщение "Несчастливый билет".

Проверим правильность работы кода:

1. Пример с номером `123456`:
   - Сумма первых трех цифр: 1 + 2 + 3 = 6
   - Сумма последних трех цифр: 4 + 5 + 6 = 15
   - Суммы не совпадают, поэтому выводится "Несчастливый билет".
   - Результат правильный.
   
2. Пример с номером `123321`:
   - Сумма первых трех цифр: 1 + 2 + 3 = 6
   - Сумма последних трех цифр: 3 + 2 + 1 = 6
   - Суммы совпадают, поэтому выводится "Счастливый билет".
   - Результат правильный.

Таким образом, код корректно определяет, является ли заданный номер билета счастливым или нет.

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:  
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`  

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

### Решение Задания 6

1. Для выполнения данного задания будем использовать конструкциию `if-elif-else`
2. Модуль `math`

````python
import math # Подключение модуля math
# Запрашиваем тип фигуры у пользователя
type = input("Введите тип фигуры: ")
# Если тип фигуры = "Треугольник"
if type == "Треугольник":
    # Запрашиваем длины сторон треугольника у пользователя
    a = float(input("Введите длину стороны A: ").replace(',', '.'))
    b = float(input("Введите длину стороны B: ").replace(',', '.'))
    c = float(input("Введите длину стороны C: ").replace(',', '.'))
    # Вычисляем полупериметр
    p = (a + b + c) / 2
    # Вычисляем площадь треугольника
    area = round(math.sqrt(p * (p - a) * (p - b) * (p - c)), 2)
    # Выводим площадь треугольника
    print("Площадь треугольника:", area)
# Если тип фигуры = "Круг"
elif type == "Круг":
    # Запрашиваем радиус круга у пользователя
    radius = float(input("Введите радиус круга: ").replace(',', '.'))
    # Вычисляем площадь круга
    area = round(math.pi * radius**2, 2)
    # Выводим площадь круга
    print("Площадь круга:", area)
# Если тип фигуры = "Прямоугольник"
elif type == "Прямоугольник":
    # Запрашиваем длины сторон прямоугольника у пользователя
    a = float(input("Введите длину стороны A: ").replace(',', '.'))
    b = float(input("Введите длину стороны B: ").replace(',', '.'))
    # Вычисляем площадь прямоугольника1
    area = round(a * b, 2)
    # Выводим площадь прямоугольника
    print("Площадь прямоугольника:", area)
````

**Описание работы кода:**

Данный код предлагает пользователю ввести тип фигуры ("Треугольник", "Круг" или "Прямоугольник") и, в зависимости от выбранного типа, запрашивает соответствующие данные (длины сторон треугольника, радиус круга или длины сторон прямоугольника). Затем код вычисляет площадь фигуры и выводит ее.

Разберем подробнее каждую часть кода:

1. Запрос типа фигуры у пользователя:
   - Переменная type получает введенное значение типа фигуры с помощью input().

2. Вычисление площади треугольника:
   - Если type равно "Треугольник", код просит пользователя ввести длины сторон треугольника.
   - Затем вычисляется полупериметр треугольника с помощью формулы (a + b + c) / 2 и сохраняется в переменной p.
   - Вычисляется площадь треугольника с помощью формулы Heron и сохраняется в переменной area.
   - Затем площадь треугольника выводится на экран с помощью print().

3. Вычисление площади круга:
   - Если type равно "Круг", код просит пользователя ввести радиус круга.
   - Вычисляется площадь круга с помощью формулы pi * radius^2 и сохраняется в переменной area.
   - Затем площадь круга выводится на экран с помощью print().

4. Вычисление площади прямоугольника:
   - Если type равно "Прямоугольник", код просит пользователя ввести длины сторон прямоугольника.
   - Вычисляется площадь прямоугольника с помощью формулы a * b и сохраняется в переменной area.
   - Затем площадь прямоугольника выводится на экран с помощью print().

В итоге, в зависимости от выбранного типа фигуры, код вычисляет площадь треугольника, круга или прямоугольника и выводит ее на экран.
