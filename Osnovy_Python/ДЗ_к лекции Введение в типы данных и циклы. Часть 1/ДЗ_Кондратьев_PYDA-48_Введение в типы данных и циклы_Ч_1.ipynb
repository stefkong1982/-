## Задание 1
Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:
- среднюю букву, если число букв в слове нечетное;
- две средних буквы, если число букв четное.

Примеры работы программы:  
1. `word = 'test'`

Результат:  
`es`  
  
2.     
`word = 'testing'`

Результат:  
`t`

### Решение Задание 1

**Описание решения**

Для решения задачи будем использовать

1. Конструкцию `start:stop:step`
2. Оператор целочисленного деления (`//`) для определения центральной позиции в строке. Если длина строки нечетная, мы возвращаем символ в этой позиции. Если длина строки четная, мы возвращаем два символа, используя срез строки.
3. Конструкциию `if-elif-else` и оператор `[]`




````python
# @title Код решения
word = input("Введите слово из латинских букв: ")
index = len(word) // 2

if len(word) % 2 == 0:
    print(word[index - 1 : index + 1])
else:
    print(word[index])
````


Для решения данного задания будем использовать цикл `while`,

````python
# @title Код решения
number = int(input("Введите число: "))
sum_numbers = 0

while number != 0:
    sum_numbers += number
    number = int(input("Введите число: "))
print("Результат:", sum_numbers)
````

## Задание 4

У нас есть список, содержащий информацию о среднедневной температуре в Фаренгейтах за произвольный период по странам (структура данных в примере).
Необходимо написать код, который рассчитает среднюю температуру за период в Цельсиях(!) для каждой страны.

Пример работы программы:  
```
countries_temperature = [
    ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Германия', [57.2, 55.4, 59, 59, 53.6]],
    ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]
```

Результат:  
```
Средняя температура в странах:
Таиланд  -  23.9 С
Германия  -  13.8 С
Россия  -  3.7 С
Польша  -  12.0 С
```

### Решение Задание 4

Температуру в градусах Фаренгейта можно перевести в градусы Цельсия, используя следующую формулу:

C = (F - 32) * 5/9

где
C - это температура в градусах Цельсия,
F - это температура в градусах Фаренгейта.

Например, если вы хотите перевести 70 градусов Фаренгейта в Цельсия, вы просто вставите 70 вместо F в формулу:

C = (70 - 32) * 5/9 = 21.11

Таким образом, 70 градусов Фаренгейта равны приблизительно 21.11 градусов Цельсия.

````python
# @title Код решения
countries_temperature = [
    ['Таиланд', [75.2, 77, 78.8, 73.4, 68, 75.2, 77]],
    ['Германия', [57.2, 55.4, 59, 59, 53.6]],
    ['Россия', [35.6, 37.4, 39.2, 41, 42.8, 39.2, 35.6]],
    ['Польша', [50, 50, 53.6, 57.2, 55.4, 55.4]]
]


print("Средняя температура в странах:")
for country_info in countries_temperature:
    country = country_info[0]
    temp_fahrenheit = country_info[1]
    average_temp_fahrenheit = sum(temp_fahrenheit) / len(temp_fahrenheit)
    average_temp_celsius = (average_temp_fahrenheit - 32) * 5 / 9
    print(f"{country} - {round(average_temp_celsius, 1)} °C")
````

**Описание работы кода:**

1. `print("Средняя температура в странах:")` - эта строка просто выводит заголовок на экран.

2. `for country_info in countries_temperature:` - эта строка начинает цикл, который пройдет по всем элементам списка `countries_temperature`. Каждый элемент этого списка - это подсписок, который содержит название страны и список средних дневных температур в Фаренгейтах.

3. `country = country_info[0]` - это строка извлекает название страны из текущего подсписка и сохраняет его в переменной `country`.

4. `temp_fahrenheit = country_info[1]` - аналогично предыдущей строке, это строка извлекает список температур из текущего подсписка и сохраняет его в переменной `temp_fahrenheit`.

5. `average_temp_fahrenheit = sum(temp_fahrenheit) / len(temp_fahrenheit)` - эта строка вычисляет среднюю температуру в Фаренгейтах, подсчитывая сумму всех температур и делая ее на количество этих температур.

6. `average_temp_celsius = (average_temp_fahrenheit - 32) * 5 / 9` - эта строка переводит среднюю температуру из Фаренгейтов в Цельсии по формуле для перевода температуры из в Фаренгейтах в Цельсии.

7. `print(f"{country} - {round(average_temp_celsius, 1)} °C")` - эта строка выводит на экран название страны и ее среднюю температуру в Цельсиях, округленную до первого знака после запятой.

## Задание 5 (необязательное)

Имеется список с транспортными номерами. Необходимо написать код, который проверит каждый номер на валидность (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то вывести его в нужном формавте (пример ниже), а если не валиден — вывести текст. При решении помогут регулярные выражения, с которыми будет знакомство на практике.

Примеры работы программы:  

`car_ids = ['А222ВС96', 'АБ22ВВ193']`

Результат:
```
Номер А222ВС валиден. Регион: 96
Номер АБ22ВВ193 не валиден
```

### Решение Задание 5

В автомобильных номерах России используются только 12 букв кириллического алфавита, которые по форме схожи с буквами латинского алфавита:
```
А, В, Е, К, М, Н, О, Р, С, Т, У, Х.
```
Это сделано для упрощения международного общения и предотвращения недоразумений при расшифровке номеров.

Например, символы "А" или "В" будут одинаково восприниматься как на Кириллице, так и на Латинице. При этом из-за отсутствия таких символов на латинице, как "Ы" или "Э", эти буквы не используются в автомобильных номерах, чтобы избежать путаницы или затруднений при чтении номера в странах, где используется латинский алфавит. Это сделано в соответствии с конвенцией ООН.

Также стоит учесть важность распознавания номеров на международном уровне в рамках борьбы с преступностью, когда машины могут пересекать границы стран, и должна быть возможность легко идентифицировать номер автомобиля.

Для решения будем использовать `регулярные выражения` и функцию `re.findall()`

````python
# @title Код решения
import re

car_ids = ['А222ВС96', 'АБ22ВВ193', 'О777ОО777']
pattern = r'[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}'

for car_id in car_ids:
    if re.findall(pattern, car_id):
        print(f'Номер {car_id[:6]} валиден. Регион: {car_id[6:]}')
    else:
        print(f'Номер {car_id} не валиден')
````

**Описание работы кода:**

Этот код использует `регулярные выражения`, чтобы проверить, соответствуют ли указанные номера автомобилей определенной модели. Регулярное выражение, указанное в шаблоне `pattern`, ищет следующую структуру в строке:

1. `[АВЕКМНОРСТУХ]` - один символ из перечисленного набора. Это буквы, которые используются в русских номерах автомобилей.

2. `\d{3}` - три цифры (символ `d` представляет любую цифру, а `{3}` указывает, что мы ищем ровно три из них подряд).

3. `[АВЕКМНОРСТУХ]{2}` - две буквы из перечисленного набора, идущие подряд.

4. `\d{2,3}` - от двух до трех цифр подряд. Это представляет регион.

Этот код проходит по каждому ID в списке `car_ids` и выясняет, соответствует ли каждый элемент этому шаблону. В случае совпадения он выводит сообщение, говорящее, что номер валиден и указывает его регион. Иначе говорит, что номер не валиден.

## Задание 6 (необязательное)

Дан поток логов по количеству просмотренных страниц для каждого пользователя (`пользователь,дата;просмотры`). Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя.
Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей. И тут регулярные выражения облегчат немного реализацию.

Примеры работы программы:

### Решение Задание 6
Для решения будем использовать регулярные выражения и функцию `re.match()`

````python
# @title Код решения
import re

stream = [
    'user4,2021-01-01;3',
    'user3,2022-01-07;4',
    'user2,2022-03-29;1',
    'user1,2020-04-04;13',
    'user2,2022-01-05;7',
    'user1,2021-06-14;4',
    'user3,2022-07-02;10',
    'user4,2021-03-21;19',
    'user4,2022-03-22;4',
    'user4,2022-04-22;8',
    'user4,2021-05-03;9',
    'user4,2022-05-11;11'
]

views_per_user = {}
unique_users = set()

for log in stream:
    match_result = re.match(r'(.*),.*;(\d+)', log)
    user = match_result.group(1)
    views = int(match_result.group(2))
    if user in views_per_user:
        views_per_user[user] += views
    else:
        views_per_user[user] = views
    unique_users.add(user)

average_views_per_user = sum(views_per_user.values()) / len(unique_users)

print("Среднее количество просмотров на уникального пользователя:", average_views_per_user)
````

**Описание работы кода**

Сначала мы импортируем модуль `re`, который предоставляет поддержку регулярных выражений в Python.

Затем у нас объявлен список `stream`, который включает данные о пользователях, дате посещения и количестве просмотров страниц.

Мы создаем две пустые структуры данных: словарь `views_per_user` для хранения общего количества просмотров для каждого пользователя и множество `unique_users` для хранения уникальных пользователей.

Затем начинается цикл `for`, в котором каждая строка в списке `stream` обрабатывается.
Первым делом, мы используем регулярное выражение, чтобы разделить и извлечь имя пользователя и количество просмотров из каждой строки. Регулярное выражение `(.*),.*;(\d+)` ищет две группы символов, разделенных запятой и точкой с запятой. Первая группа `(.*),` соответствует любому количеству любых символов, оканчивающихся запятой, вторая `\d+` соответствует одной или нескольким цифрам.

Фрагмент `.*` соответствует любому символу (кроме новой строки), повторенному ноль или более раз. Символ `;` - это именно тот знак, который она должна найти - точку с запятой. Таким образом, часть `.*;` извлекает все символы между запятой и точкой с запятой, игнорируя их при сохранении результатов.

Поэтому, когда мы получаем `match_result.group(1)` и `match_result.group(2)`, мы фактически получаем только имя пользователя и количество просмотров.

Вся остальная часть строки (дата и символы `,` и `;`) игнорируется, поскольку в регулярном выражении они находятся вне скобок `(` `)`, которые обозначают группы для сохранения результатов.

Затем извлекается имя пользователя и количество просмотров, преобразованное в целое число.

Далее проверяется, есть ли пользователь в словаре `views_per_user`. Если пользователь уже есть в словаре, то к его количеству просмотров прибавляется новое значение. Если нет - пользователь добавляется в словарь с текущим количеством просмотров.

После этого каждое уникальное имя пользователя добавляется в множество `unique_users`.

Наконец, после завершения цикла, вычисляется и выводится среднее количество просмотров на уникального пользователя. В `average_views_per_user` суммируются значения просмотров из словаря и делятся на количество уникальных пользователей.

Таким образом, данный код обрабатывает данные о пользователях и их просмотрах и вычисляет среднее количество просмотров на уникального пользователя.
