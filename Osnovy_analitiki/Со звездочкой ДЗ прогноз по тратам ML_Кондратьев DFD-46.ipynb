```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('ДЗ Google Таблицы (Кондратьев С.П. DFD-(46)(54)) - ДЗ маш. обучение.csv', sep=',')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
x = df[['траты в позапрошлом месяце', 'траты в прошлом месяце','траты в текущем месяце']]
y = df['траты в следующем месяце']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
lr = LinearRegression() 
```

```python
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.predict(x_test)
```

```python
pd.DataFrame([y_test.values, lr.predict(x_test)]).T
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, lr.predict(x_test))
```

```python
mean_squared_error(y_train, lr.predict(x_train))
```
